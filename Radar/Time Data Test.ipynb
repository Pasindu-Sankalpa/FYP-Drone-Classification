```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.fromfile("PostProc_22_1/adc_data.bin", dtype=np.uint16)
data = data.reshape((128, 128, 256, 8))[:, :, :, [0, 4]]
data = data - (data>=np.power(2, 15))*np.power(2, 16)
beat = (data[:, :, :, 0] + 1j*data[:, :, :, 1]).astype(np.complex64)
```

```python
beat_frame = beat[0]

for chirp in beat_frame[:3]:
    plt.plot(np.abs(chirp))
```
